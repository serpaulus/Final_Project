```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the XLS file as a Pandas DataFrame and preview the DataFrame:
df_table = pd.read_csv(r'C:\Users\Sergio_Paramo\Documents\Projects\Domino\DomnoBaseline\Ph_1_downloads\T_ITEM_102721.csv', low_memory=False)
df_table
```

Total number of rows 

```python
rows = len(df_table)
rows
```

```python
# Check table data types 
data_type = pd.DataFrame(
    df_table.dtypes,
    columns=['Data Type']
)
```

```python
# Check missing data
missing_data = pd.DataFrame(
    df_table.isnull().sum(), 
    columns=['Missing Values']
)

missing_data
```

```python
# Unique value check
unique_values = pd.DataFrame(
    columns=['Unique Values']
)
for row in list(df_table.columns.values):
    unique_values.loc[row] = [df_table[row].nunique()]
```

```python
# Column Compleness % 
comp = pd.DataFrame(
    columns=['% Complete']
)
for row in list(df_table.columns.values):
    comp.loc[row] = [df_table[row].nunique()*100/rows]    
comp
```
