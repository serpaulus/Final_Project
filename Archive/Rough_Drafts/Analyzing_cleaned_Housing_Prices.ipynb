```python
sqft_df = clean_train_df[['LivingArea_Sqft','1stFloor_Sqft','2ndFloor_Sqft', 'Sale_Price']]
sqft_df = sqft_df[~sqft_df['Sale_Price'].isnull()]
sqft_vars = list(sqft_df)
del sqft_vars[-1]

sqft_summary = []

for sqft_type in sqft_vars:
    corr_val = np.corrcoef(sqft_df[sqft_type], sqft_df['Sale_Price'])[1][0]
    sqft_summary.append(corr_val)
    
pd.DataFrame([sqft_summary], columns=sqft_vars, index=['SalePrice Correlation'])
```
